### 🧪 Quest 1: The Math Module Rune Circle

````python
# Quest 1: The Math Module Rune Circle
import math

radius = 7
print("=== MATH MODULE RUNES ===")
print(f"Circle radius        : {radius}")
print(f"Circle Area (pi*r^2) : {math.pi * radius ** 2:.2f}")
print(f"Square root of 144   : {math.sqrt(144)}")
print(f"Ceiling of 4.2       : {math.ceil(4.2)}")
print(f"Floor of 4.8         : {math.floor(4.8)}")
print(f"Factorial of 6       : {math.factorial(6)}")
print(f"GCD of 48 and 18     : {math.gcd(48, 18)}")
print(f"log base 2 of 1024   : {math.log2(1024)}")
````


---
### 🧪 Quest 2: The Bitwise Rune Gate

> *A gate covered in binary glyphs will only open for those who understand `&`, `|`, `^`, `~`, `<<`, and `>>` — the hidden language beneath every number.*


````python
# Quest 2: The Bitwise Rune Gate — hidden binary magic
x, y = 12, 10  # 1100 and 1010 in binary

print("=== BITWISE RUNE GATE ===")
print(f"x = {x} ({bin(x)})")
print(f"y = {y} ({bin(y)})")
print(f"x & y  (AND)         : {x & y}  ({bin(x & y)})")
print(f"x | y  (OR)          : {x | y}  ({bin(x | y)})")
print(f"x ^ y  (XOR)         : {x ^ y}  ({bin(x ^ y)})")
print(f"~x     (NOT)         : {~x}")
print(f"x << 2 (Left Shift)  : {x << 2}  — doubles power twice")
print(f"x >> 2 (Right Shift) : {x >> 2}  — halves power twice")
````

### 🧪 Quest 3: The Membership & Identity Wards

> *Two guardian wards stand at the threshold: one asks "do you belong?" (`in` / `not in`), the other asks "are you truly the same, or just a copy?" (`is` / `==`).*


````python
# Quest 3: The Membership & Identity Wards
inventory = ["sword", "shield", "potion", "map"]

print("=== MEMBERSHIP & IDENTITY WARDS ===")
print(f"Inventory: {inventory}")
print(f"'sword' in inventory       : {'sword' in inventory}")
print(f"'bow' in inventory         : {'bow' in inventory}")
print(f"'bow' not in inventory     : {'bow' not in inventory}")

a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f"\na = {a}, b = {b}, c = a")
print(f"a == b (same values?)      : {a == b}")
print(f"a is b (same object?)      : {a is b}")
print(f"a is c (same object?)      : {a is c}")
````

### 🧪 Quest 4: The PEMDAS Puzzle Lock

> *An ancient lock etched with `5 + 3 * 2 ** 2 - (8 / 4)` will only open if you can trace the correct order of operations: Parentheses → Exponents → Multiply/Divide → Add/Subtract.*


````python
# Quest 4: Order of Operations — The PEMDAS Puzzle Lock
puzzle = 5 + 3 * 2 ** 2 - (8 / 4)

print("=== THE PEMDAS PUZZLE LOCK ===")
print("Expression: 5 + 3 * 2 ** 2 - (8 / 4)")
print("Step 1 — Parentheses : (8 / 4)  =  2.0")
print("Step 2 — Exponent    : 2 ** 2   =  4")
print("Step 3 — Multiply    : 3 * 4    =  12")
print("Step 4 — Add/Sub     : 5 + 12 - 2.0")
print(f"\nFinal Answer: {puzzle}")
print("The lock clicks open!" if puzzle == 15.0 else "Wrong combination.")
````


---
### ⚔️ Boss Battle: The Compound Interest Cauldron

> *The Armourer's final trial: a merchant's cauldron that grows gold over time. Calculate the treasure after years of compounding — the boss falls only to the correct formula:*
> `A = P(1 + r/n)^(nt)`


````python
# ⚔️ Boss Battle: The Compound Interest Cauldron
# A merchant offers to grow your gold over time. Calculate the final treasure.

principal = 5000        # starting gold
rate      = 0.08        # 8% annual growth enchantment
years     = 10
n         = 12          # compounded monthly

final_amount = principal * (1 + rate / n) ** (n * years)
interest_earned = final_amount - principal

print("═══════════════════════════════════")
print("   COMPOUND INTEREST CAULDRON      ")
print("═══════════════════════════════════")
print(f"  Starting Gold   : {principal}")
print(f"  Growth Rate     : {rate*100:.0f}% per year")
print(f"  Years Brewing   : {years}")
print(f"  Compounded      : {n}x per year")
print(f"  ─────────────────────────────")
print(f"  Final Treasure  : {final_amount:.2f}")
print(f"  Interest Gained : {interest_earned:.2f}")
print("═══════════════════════════════════")
print("The cauldron bubbles with golden light!")
````

---

## 🗝️ The Three Decision Gates

```python
if condition:
    # Executes when the condition is True

elif another_condition:
    # Executes if the first condition is False
    # and this condition is True

else:
    # Executes when all previous conditions are False
```

Think of the decision process as navigating through a sequence of gates:

| Statement | Purpose |
|-----------|---------|
| `if` | Tests the first condition |
| `elif` | Tests additional conditions if previous ones fail |
| `else` | Executes when none of the conditions are satisfied |

Only **one matching branch** executes during a single decision.

---

## 🌍 Real-World Applications

Conditional statements are used in almost every software system:

- 🔐 Login authentication
- 💳 ATM and banking systems
- 🛒 E-commerce discount calculations
- 🎮 Game logic and player actions
- 🚦 Traffic control systems
- 🤖 Artificial Intelligence and Machine Learning decision pipelines

Whenever a program asks **"Should I do this?"**, it is using conditional logic.

---

## 📌 Key Concepts You'll Master

- Boolean values (`True` and `False`)
- Comparison operators
- Logical operators (`and`, `or`, `not`)
- Nested `if` statements
- Multi-way branching using `elif`
- Best practices for readable decision logic

### 🧪 Quest 5: The Quadratic Oracle

> *The Oracle poses an equation: `ax² + bx + c = 0`. She demands to know: are its roots real, or do they exist only in the imaginary realm?*

The **discriminant** `b² - 4ac` reveals the truth:
- `≥ 0` → Real roots (real world)
- `< 0` → Imaginary roots (shadow realm)

`cmath` — the complex math module — handles imaginary numbers.


````python
import cmath

a, b, c = 1, 2, 5  # The Oracle's coefficients

discriminant = b**2 - 4*a*c
root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
root2 = (-b - cmath.sqrt(discriminant)) / (2*a)

if discriminant >= 0:
    print("The roots are REAL — they exist in this world!")
else:
    print("The roots are IMAGINARY — they dwell in the shadow realm!")

print(f"Root 1 : {root1}")
print(f"Root 2 : {root2}")
````

## 🔄 The Two Loop Mechanisms

Python provides two primary looping constructs, each designed for different situations.

Use a **`for` loop** when the number of iterations is known or when iterating through a collection.

```python
for i in range(5):
```

```python
hp = 100
```

These statements provide precise control over loop execution and help build efficient algorithms.

````python
# Find all Armstrong numbers between 1 and 500
print("Armstrong Numbers (1–500):")
armstrong_found = []

for num in range(1, 501):
    digits = str(num)
    n = len(digits)
    total = sum(int(d)**n for d in digits)
    if total == num:
        armstrong_found.append(num)

print(armstrong_found)
print(f"Total found: {len(armstrong_found)} Armstrong rooms unlocked!")
````

````python
def find_max(lst):
    maximum = lst[0]
    for num in lst[1:]:
        if num > maximum:
            maximum = num
    return maximum

def find_min(lst):
    minimum = lst[0]
    for num in lst[1:]:
        if num < minimum:
            minimum = num
    return minimum

numbers = [34, 12, 78, 5, 91, 23, 67]

print(f"Number pool  : {numbers}")
print(f"Maximum      : {find_max(numbers)} ")
print(f"Minimum      : {find_min(numbers)} ")
````

## 🏗️ Classes and Objects

A **class** is a blueprint that defines the structure and behavior of an object.

An **object** is an instance created from that blueprint.

```python
class Hero:

    def __init__(self, name, health):
        self.name = name
        self.health = health

    def attack(self):

player = Hero("Arthur", 100)
player.attack()
```

## 🧩 Anatomy of a Class

| Component | Purpose |
|-----------|---------|
| `class` | Defines a blueprint |
| `__init__()` | Constructor executed during object creation |
| `self` | Refers to the current object |
| Attributes | Store object data |
| Methods | Define object behavior |
| Object | Instance created from a class |

## 🏛️ The Four Pillars of OOP

| Principle | Description | Example |
|-----------|-------------|---------|
| **Encapsulation** | Bundle data and methods together while protecting internal data | Player health is managed through methods instead of direct modification |
| **Inheritance** | Create new classes from existing ones | `Paladin` inherits from `Warrior` |

````python
# ── FORGING A BASIC HERO ──────────────────────────────────────────
class Hero:
    def __init__(self, name, hp, attack):
        self.name   = name
        self.hp     = hp
        self.attack = attack

    def introduce(self):
        return f"I am {self.name} — HP: {self.hp}, ATK: {self.attack}"

    def take_damage(self, damage):
        self.hp -= damage
        if self.hp <= 0:
            return f"{self.name} has fallen!"
        return f"{self.name} took {damage} damage. HP left: {self.hp}"

# Forge two heroes
hero1 = Hero("Aria", 100, 35)
hero2 = Hero("Zeke", 80, 50)

print(hero1.introduce())
print(hero2.introduce())
print()
print(hero1.take_damage(30))
print(hero2.take_damage(80))
````

### 🧪 Quest 1: The Single Inheritance Chain

> *A Mage is a Hero who also wields spells. They inherit everything from Hero, then add their own powers.*


````python
class Hero:
    def __init__(self, name, hp):
        self.name = name
        self.hp   = hp

    def describe(self):
        return f"Hero: {self.name} | HP: {self.hp}"

class Mage(Hero):
    def __init__(self, name, hp, mana):
        super().__init__(name, hp)  # Call parent constructor
        self.mana = mana

    def cast_spell(self, spell):
        cost = 20
        if self.mana >= cost:
            self.mana -= cost
            return f"{self.name} casts {spell}! (Mana left: {self.mana})"
        return f"Not enough mana!"

    def describe(self):  # Override parent method
        return f"Mage : {self.name} | HP: {self.hp} | Mana: {self.mana}"

mage = Mage("Elara", 70, 150)
print(mage.describe())
print(mage.cast_spell("Fireball"))
print(mage.cast_spell("Ice Storm"))
print(mage.cast_spell("Blizzard"))
print(mage.take_damage(20) if hasattr(mage, 'take_damage') else "No take_damage here")
````

### 🧪 Quest 2: Multilevel Inheritance — The Legendary Line

> *Warrior → Knight → Paladin. Each generation inherits and extends the last.*


````python
class Warrior:
    def __init__(self, name):
        self.name = name
    def fight(self):
        return f"{self.name} swings a sword!"

class Knight(Warrior):
    def __init__(self, name, armor):
        super().__init__(name)
        self.armor = armor
    def defend(self):
        return f"{self.name} raises a {self.armor} shield!"

class Paladin(Knight):
    def __init__(self, name, armor, faith):
        super().__init__(name, armor)
        self.faith = faith
    def holy_strike(self):
        return f"{self.name} channels {self.faith} into a divine blow!"

p = Paladin("Sir Aldric", "Dragonscale", "Lightbringer")
print(p.fight())        # Inherited from Warrior
print(p.defend())       # Inherited from Knight
print(p.holy_strike())  # Paladin's own power
````

### 🧪 Quest 3: Multiple Inheritance — The Hybrid Class

> *Some adventurers are born of two bloodlines. The `HybridWarrior` inherits from both `Swordsman` and `Archer`.*


````python
class Swordsman:
    def melee_attack(self):
        return "Slashes with a blade!"

class Archer:
    def ranged_attack(self):
        return "Fires a precise arrow!"

class HybridWarrior(Swordsman, Archer):
    def __init__(self, name):
        self.name = name
    def combo_attack(self):
        return f"{self.name}: {self.melee_attack()} then {self.ranged_attack()}"

ranger = HybridWarrior("Voss")
print(ranger.melee_attack())
print(ranger.ranged_attack())
print(ranger.combo_attack())
print(f"MRO: {[c.__name__ for c in HybridWarrior.__mro__]}")
````

### ⚔️ Boss: Operator Overloading — The Power Combiner

> *The Forgemaster's final test: two weapons can be combined with `+`. Overload the `__add__` operator to merge their stats.*


````python
class Weapon:
    def __init__(self, name, damage, durability):
        self.name       = name
        self.damage     = damage
        self.durability = durability

    def __str__(self):
        return f"[{self.name} | DMG:{self.damage} | DUR:{self.durability}]"

    def __add__(self, other):
        # Combine two weapons into a legendary one
        combined_name = f"{self.name}-{other.name} Fusion"
        return Weapon(
            combined_name,
            self.damage + other.damage,
            (self.durability + other.durability) // 2
        )

    def __gt__(self, other):
        return self.damage > other.damage

sword  = Weapon("Shadowblade", 80, 100)
staff  = Weapon("Runestaff",   65, 120)
fusion = sword + staff

print(f"Weapon 1  : {sword}")
print(f"Weapon 2  : {staff}")
print(f"Fused     : {fusion}")
print(f"Sword stronger than Staff? {sword > staff}")
print("\nThe Forge of Beings bows to your mastery!")
````

````python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(12, 9))
fig.suptitle("THE ORACLE'S VISION — Floor 9 Dashboard", fontsize=15, fontweight='bold')

adventurers = ["Aria", "Zeke", "Voss", "Nira", "Elara"]
levels = [42, 38, 45, 31, 50]
kills  = [120, 98, 145, 67, 200]
gold   = [4500, 3200, 5800, 2100, 9000]
hp     = [380, 520, 310, 270, 490]
classes = ["Mage", "Warrior", "Archer", "Rogue", "Paladin"]
class_counts = [2, 3, 1, 4, 2]

# ── 1. BAR CHART: Levels ───────────────────────────────────────────
colors = ['#e74c3c','#3498db','#2ecc71','#f39c12','#9b59b6']
axes[0,0].bar(adventurers, levels, color=colors, edgecolor='black')
axes[0,0].set_title("Adventurer Levels")
axes[0,0].set_ylabel("Level")
axes[0,0].set_ylim(0, 60)
for i, (name, lvl) in enumerate(zip(adventurers, levels)):
    axes[0,0].text(i, lvl + 0.5, str(lvl), ha='center', fontweight='bold')

# ── 2. LINE CHART: Gold Earned ─────────────────────────────────────
axes[0,1].plot(adventurers, gold, marker='o', color='gold',
               linewidth=2.5, markersize=10, markerfacecolor='darkgoldenrod')
axes[0,1].fill_between(range(len(adventurers)), gold, alpha=0.15, color='gold')
axes[0,1].set_title("Gold Accumulated")
axes[0,1].set_ylabel("Gold Coins")
axes[0,1].set_xticks(range(len(adventurers)))
axes[0,1].set_xticklabels(adventurers)

# ── 3. SCATTER: HP vs Kills ────────────────────────────────────────
scatter = axes[1,0].scatter(hp, kills, c=levels, cmap='plasma',
                             s=200, edgecolors='black', linewidth=1.5)
plt.colorbar(scatter, ax=axes[1,0], label='Level')
axes[1,0].set_title("HP vs Kills (color = level)")
axes[1,0].set_xlabel("HP")
axes[1,0].set_ylabel("Kills")
for i, name in enumerate(adventurers):
    axes[1,0].annotate(name, (hp[i]+5, kills[i]+2), fontsize=8)

# ── 4. PIE CHART: Class Distribution ──────────────────────────────
explode = [0.05]*5
axes[1,1].pie(class_counts, labels=classes, colors=colors,
              autopct='%1.0f%%', explode=explode, startangle=90)
axes[1,1].set_title("Class Distribution")

plt.tight_layout()
plt.savefig("oracle_vision.png", dpi=120, bbox_inches='tight')
plt.show()
print("\nThe Oracle's Vision has been recorded!")
````
